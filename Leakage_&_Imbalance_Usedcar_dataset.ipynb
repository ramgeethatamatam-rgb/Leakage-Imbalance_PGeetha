Task 2:Clean the Data Fix the issues identified: drop null target rows, impute missing input features, strip and lowercase the brand column, extract numeric values from the mileage column, and remove duplicates.

```python
#cleaning the Brand values



df['brand'] = df ['brand'].str.strip().str.lower()
```

```python
print(f"Unique values before cleaning: {df['brand'].nunique()}")

df['brand'] = df['brand'].str.strip()
print(f"Unique values After cleaning: {df['brand'].nunique()}")

print(df['brand'].value_counts())
```

```python
df
```

```python
df['mileage'] = df['mileage'].astype(str).str.replace('kmpl', '').str.replace('kmp', '').str.strip()

df['mileage'] = pd.to_numeric(df['mileage'], errors='coerce')
df['mileage'].head()
```

Task 3 — Compute Baseline MAE Build a baseline model that predicts the mean selling_price for every record. Calculate and print the MAE of this baseline on the full cleaned dataset.

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

mean_selling_price = df['selling_price'].mean()
mean_selling_price

y_true = df['selling_price']
y_pred = np.full(len(df) ,mean_selling_price)

mae = mean_absolute_error(y_true, y_pred)

print("Baseline MAE:",mae)
```
